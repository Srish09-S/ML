```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

penguins = sns.load_dataset("penguins")

penguins = penguins.dropna()
```

```python
sns.scatterplot(data=penguins,
                x="body_mass_g",
                y="flipper_length_mm",
                hue="species")
plt.show()
```

```python
sns.relplot(data=penguins,
                x="body_mass_g",
                y="flipper_length_mm",
                hue="species")
plt.show()
```

```python
sns.lineplot(data=penguins,
             x="island",
             y="body_mass_g",
             estimator="mean",
             errorbar="sd")
plt.show()
```

```python
sns.lineplot(data=penguins,
             x="island",
             y="body_mass_g",
             estimator="mean",
             errorbar="sd")
plt.show()
```

```python
sns.relplot(data=penguins,
            x="island",
            y="body_mass_g",
            kind="line",
            estimator="mean",
            errorbar="sd")
plt.show()
```
