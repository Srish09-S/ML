```python
import seaborn as sns
import plotly.express as px
```

```python
country_df = px.data.gapminder()
country_df
```

```python
temp_df = country_df[country_df['continent'] == 'Asia'].pivot(index = 'country' , columns = 'year' , values = 'lifeExp')
temp_df
     
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(temp_df) 
```
